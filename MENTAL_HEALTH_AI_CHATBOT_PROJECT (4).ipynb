```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

path ="/content/Student Mental health.csv"
data = pd.read_csv(path)
data
```

```python
data.shape
```

```python
data.tail()
```

```python
data.isna().sum()
```

```python
data=data.dropna()
#this is used to clean data (to remove missing information from the data)
```

```python
data.isna().sum()
#By this we can conclude our data colums has no missing information
```

```python
#convert words to numbers(Encoding)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# This simple loop turns every text column into a number automatically
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])

data.head()
 # Run this to see the numbers!
```

```python
from sklearn.model_selection import train_test_split

# X is the information (Age, Gender, etc.). y is what we want to predict (Anxiety).
X = data.drop(['Do you have Anxiety?', 'Timestamp'], axis=1)
y = data['Do you have Anxiety?']

# We give 80% to study and keep 20% for the final test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Create the brain (MLP) and tell it to learn (fit)
mlp = MLPClassifier(hidden_layer_sizes=(10, 5), max_iter=2000, random_state=1)
mlp.fit(X_train, y_train)

# Take the final exam and see the score
predictions = mlp.predict(X_test)
print("My Chatbot Accuracy is:", accuracy_score(y_test, predictions))
```

```python
import joblib
from google.colab import files

# Save the model
joblib.dump(mlp, 'mental_health_model.pkl')
print(" Model saved!")

# Save the label encoder
joblib.dump(le, 'label_encoder.pkl')
print("Label encoder saved!")

# Download both files
files.download('mental_health_model.pkl')
files.download('label_encoder.pkl')
print("Files downloading to your computer!")
```
